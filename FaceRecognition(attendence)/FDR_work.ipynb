```python
import face_recognition

# Load the image with faces
image_path = "tony1.jpg"
image = face_recognition.load_image_file(image_path)

# Find all face locations in the image
face_locations = face_recognition.face_locations(image)

# Check if at least one face is found in the image
if len(face_locations) > 0:
    # Compute face embeddings for all detected facespip
    face_encodings = face_recognition.face_encodings(image, face_locations)

    # The face_encodings list contains the face embeddings as numpy arrays
    for i, face_encoding in enumerate(face_encodings):
        print(f"Face {i + 1} embedding shape: {face_encoding.shape}")

else:
    print("No faces found in the image.")
```

```python
import face_recognition
import pandas as pd
import numpy as np

# Load the image with faces
image_path = "fahad3.jpg"
image = face_recognition.load_image_file(image_path)

# Find all face locations in the image
face_locations = face_recognition.face_locations(image)

# Check if at least one face is found in the image
if len(face_locations) > 0:
    # Compute face embeddings for all detected faces
    face_encodings = face_recognition.face_encodings(image, face_locations)

    # Convert the face embeddings to a list of strings
    face_encodings_str = [", ".join(map(str, encoding)) for encoding in face_encodings]

    # Create a list of dictionaries where each dictionary represents a face and its embedding
    faces_data = [{"emp_id": i + 1, "Embedding": encoding} for i, encoding in enumerate(face_encodings_str)]

    # Create a Pandas DataFrame from the list of dictionaries
    df = pd.DataFrame(faces_data)

    # Save the DataFrame to a CSV file
    df.to_csv('check_embedding.csv', index=False)

    print("Face embeddings saved to 'face_embeddings.csv'.")

else:
    print("No faces found in the image.")
```
